```python
import pandas as pd

# Load the first CSV file
file1_path = '/var/tmp/PredictiveManteinanceEngineTraining.csv'
df1 = pd.read_csv(file1_path)

# Load the second CSV file
file2_path = '/var/tmp/PredictiveManteinanceEngineValidation.csv'
df2 = pd.read_csv(file2_path)

# Display the first few rows of each DataFrame to verify
print("First CSV file:")
print(df1.head())

print("\nSecond CSV file:")
print(df2.head())
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

# Assuming the datasets are already loaded as df1 (training) and df2 (validation/testing)
train_df = df1
test_df = df2

# Display the first few rows of each DataFrame to verify
print("Training Data:")
print(train_df.head())

print("\nTesting Data:")
print(test_df.head())

# Select the features and target variable from the training set
X_train = train_df.drop(['RUL', 'label1', 'label2', 'cycle_norm'], axis=1)
y_train = train_df['RUL']

# Select the features and target variable from the testing set
X_test = test_df.drop(['RUL', 'label1', 'label2', 'cycle_norm'], axis=1)
y_test = test_df['RUL']

# Initialize and train the Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Get feature importances
feature_importances = rf.feature_importances_
features = X_train.columns

# Create a DataFrame for visualization
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Plot feature importances
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.gca().invert_yaxis()
plt.show()

# Print the feature importances
print(importance_df)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
import matplotlib.pyplot as plt

# Assuming the datasets are already loaded as df1 (training) and df2 (validation/testing)
train_df = df1
test_df = df2

# Display the first few rows of the training DataFrame to verify
print("Training Data:")
print(train_df.head())

print("\nTesting Data:")
print(test_df.head())

# Select the features and target variable from the training set
X_train = train_df.drop(['RUL', 'label1', 'label2', 'cycle_norm'], axis=1)
y_train = train_df['RUL']

# Initialize the Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Initialize RFE with the Random Forest model and select top k features
# Here, we're selecting half of the features to get a ranking
n_features_to_select = X_train.shape[1] // 2  # number of features to select
rfe = RFE(estimator=rf, n_features_to_select=n_features_to_select)
rfe.fit(X_train, y_train)

# Get the ranking of features
ranking = rfe.ranking_
features = X_train.columns

# Create a DataFrame for visualization
rfe_df = pd.DataFrame({'Feature': features, 'Ranking': ranking})
rfe_df = rfe_df.sort_values(by='Ranking')

# Plot the feature rankings
plt.figure(figsize=(10, 8))
plt.barh(rfe_df['Feature'], rfe_df['Ranking'])
plt.xlabel('RFE Ranking')
plt.ylabel('Feature')
plt.title('Feature Ranking using RFE')
plt.gca().invert_yaxis()
plt.show()

# Print the RFE rankings
print(rfe_df)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Assuming the datasets are already loaded as df1 (training) and df2 (validation/testing)
train_df = df1
test_df = df2

# List of top 12 features
top_12_features = [
    'id', 's15', 's14', 's13', 's12', 's11', 's9', 's7', 's8', 's21', 'cycle', 's2'
]

# Select the features and target variable from the training set
X_train = train_df[top_12_features]
y_train = train_df['RUL']

# Select the features and target variable from the testing set
X_test = test_df[top_12_features]
y_test = test_df['RUL']

# Initialize and train the Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Display feature importances for the selected features
feature_importances = rf.feature_importances_
importance_df = pd.DataFrame({'Feature': top_12_features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Print the feature importances
print("Feature Importances:")
print(importance_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Select the top 12 features
top_12_features = [
    'id', 's15', 's14', 's13', 's12', 's11', 's9', 's7', 's8', 's21', 'cycle', 's2'
]

# Prepare the training and validation data
X_train = train_df[top_12_features].values
y_train = train_df['RUL'].values
X_test = test_df[top_12_features].values
y_test = test_df['RUL'].values

# Normalize the data
scaler_X = MinMaxScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

scaler_y = MinMaxScaler()
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).flatten()
y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).flatten()

# Reshape data for LSTM (samples, time steps, features)
def reshape_for_lstm(data, time_steps=1):
    reshaped_data = []
    for i in range(len(data) - time_steps + 1):
        reshaped_data.append(data[i:i + time_steps])
    return np.array(reshaped_data)

# Set time steps (number of previous time steps to use)
time_steps = 10

# Reshape the data
X_train_lstm = reshape_for_lstm(X_train_scaled, time_steps)
X_test_lstm = reshape_for_lstm(X_test_scaled, time_steps)
y_train_lstm = y_train_scaled[time_steps-1:]  # Align y with the reshaped X
y_test_lstm = y_test_scaled[time_steps-1:]

# Build LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]), return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train_lstm, y_train_lstm, epochs=50, batch_size=32, validation_split=0.2)

# Evaluate the model
loss = model.evaluate(X_test_lstm, y_test_lstm)
print(f"Test Loss: {loss}")

# Make predictions
y_pred_scaled = model.predict(X_test_lstm)
y_pred = scaler_y.inverse_transform(y_pred_scaled)

# Compare predictions with true values
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(y_test, label='True Values')
plt.plot(y_pred, label='Predicted Values')
plt.xlabel('Samples')
plt.ylabel('RUL')
plt.title('True vs Predicted RUL')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
import shap
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV, train_test_split
import matplotlib.pyplot as plt

# Select the top 12 features
top_12_features = [
    'id', 's15', 's14', 's13', 's12', 's11', 's9', 's7', 's8', 's21', 'cycle', 's2'
]

# Load the data (assuming train_df and test_df are already loaded DataFrames)
# Replace with your actual data loading code if necessary

# Prepare the training and validation data
X = pd.concat([train_df[top_12_features], test_df[top_12_features]])
y = pd.concat([train_df['RUL'], test_df['RUL']])

# Normalize the data
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1)).flatten()

# Split data for training and validation
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Initialize the XGBoost model
model = xgb.XGBRegressor(objective='reg:squarederror')

from sklearn.model_selection import RandomizedSearchCV

# Define a smaller grid or randomized search space
param_grid = {
    'max_depth': [3, 6],
    'learning_rate': [0.1],
    'n_estimators': [50, 100],
    'subsample': [0.8],
    'colsample_bytree': [0.8]
}

# Use RandomizedSearchCV instead of GridSearchCV to speed things up
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid,
                                   n_iter=10, scoring='neg_mean_squared_error', cv=3,
                                   n_jobs=-1, random_state=42, verbose=2)

# Fit the randomized search to the training data
random_search.fit(X_train, y_train)

# Get best parameters and model
best_params = random_search.best_params_
best_model = random_search.best_estimator_

print("Best Parameters: ", best_params)


# Predict on the validation set
y_pred_scaled = best_model.predict(X_val)

# Inverse transform the predictions and true values
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_val_inverse = scaler_y.inverse_transform(y_val.reshape(-1, 1))

# Evaluate the model
mse = mean_squared_error(y_val_inverse, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_val_inverse, y_pred)

print(f"Validation Mean Squared Error: {mse}")
print(f"Validation Root Mean Squared Error: {rmse}")
print(f"Validation R^2 Score: {r2}")

# SHAP analysis
# Create a SHAP explainer
explainer = shap.TreeExplainer(best_model)

# Select a sample input for prediction and explanation
sample_index = 0  # You can change this to any index from X_val
sample_input = X_val[sample_index].reshape(1, -1)  # Select a single sample and reshape

# Get SHAP values for the selected sample
shap_values = explainer.shap_values(sample_input)

# Predict the RUL for the selected sample
sample_pred_scaled = best_model.predict(sample_input)
sample_pred = scaler_y.inverse_transform(sample_pred_scaled.reshape(-1, 1))

# Print predicted RUL for the sample
print(f"Predicted RUL for the selected sample: {sample_pred[0][0]}")

# Plot the SHAP values for the sample
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[0], sample_input[0], feature_names=top_12_features)
```
